#### Pandas is a Python library used for working with data sets. It has functions for analyzing, cleaning, exploring, and manipulating data.

#### We start of by importing pandas library aliased as pd.

```python
import pandas as pd
```

#### Up next we import (read) the data.
#### We use .head() which is a function that gives us an subset of the data set of the first 5 rows and the columns of the data set.

```python
df = pd.read_csv('./kc_house_data.csv')
df.head()
```

```python
df.info()
```

#### Sometimes we have so many columns that we can't view them all. Using the option_context we can view all columns.

```python
with pd.option_context('display.max_rows', 15, 'display.max_columns', None):
    display(df)
```

#### Our main target in analyzing this data set is to to check the factors affecting the price, we can start by narrowing the dataset into what seem to be affecting the price.

```python
df_1 = df[['id','price','bedrooms','bathrooms','sqft_living','waterfront','view','condition','grade','zipcode']]
df_1.head()
```

#### We will use .loc method now in order to apply conditions on our data set.

```python
df_1.loc[df_1['condition']>3]
```

#### We can narrow our vision by limiting our columns with loc.

```python
df_1.loc[df_1['condition']>3,['id','price','condition']]
```

#### Using groupby. Grouping by waterfront in order to find the average of houses with and without water front.
#### Next we are going to group by grade and condtion.

```python
df_2 = df_1.copy()
df_2[['waterfront','price']].groupby('waterfront').mean().round()
```

```python
df_2[['price','condition']].groupby('condition').mean().round()
```

```python
df_2[['price','grade']].groupby('grade').mean().round()
```

```python
df_2[['price','view']].groupby('view').mean().round()
```

```python
df_3 = df_1.copy()
df_3 = df_3[["price", "sqft_living"]]
df_3["sqft_living_range"] = pd.cut(df_3["sqft_living"],bins=10)
df_3[["price","sqft_living_range"]].groupby("sqft_living_range").mean().round().reset_index()
```

#### By isolating the factors with the price, we can see the significant effect each has on the price.

#### We can also check multiple of factors affecting the price with adding the conditoons using loc.

```python
df_2[['waterfront','view','price']].groupby(['waterfront','view']).mean().round()
```

```python
df_2[['waterfront','grade','price']].groupby(['waterfront','grade']).mean().round()
```

#### We can see the effect of having a waterfront on the grade as the minimum grade of a house with waterfront is 5

```python
df_2['special_category'] = df_2.apply(lambda x: 1 if (x['condition'] > 3
                                                       and x['waterfront'] == 1 
                                                       and x['grade'] > 7 
                                                       and x['condition'] > 3) else 0, axis=1)
df_2.loc[ df_2['bedrooms']>3,['price','special_category']].groupby('special_category').agg(['mean','max','min','count']).round()
```

#### Here we are kind of saying that we got a special request to see the house with such factors so that we have an estimation and a range of what we can expect from such a request.

```python
df_3 = df.copy()
df_3 = df_3[["price", "lat", "long"]]
df_3["lat_range"] = pd.cut(df_3["lat"], bins=8)
df_3["long_range"] = pd.cut(df_3["long"], bins=7)
df_final=df_3[["price","lat_range","long_range"]].groupby(["lat_range","long_range"]).mean().round()
df_final.dropna().sort_values('price')
```

#### Here we didn't get a very accurate result because we are kind trying to range ranges, but we can still see the effect of long and lat on the price.
